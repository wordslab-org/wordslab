```python
import numpy as np
import pandas as pd
pd.options.display.max_rows = 100
pd.options.display.max_columns = 50
```

```python
def list_datasets():
    return pd.read_csv(rootdir / "datasets.csv",sep=';')
    
datasetsdf = list_datasets()
datasetsdf
```

```python
def read_dataset_file(datasetfile):
    datasetdffile = rootdir / "datasets" / (datasetfile+".dataset.feather")
    if(datasetdffile.exists()):
        datasetdf = pd.read_feather(datasetdffile)
        print(f"Loaded dataframe for dataset {datasetfile} : {len(datasetdf)} text blocks")
        return datasetdf
    else:
        raise Exception(f"No dataframe for dataset {datasetfile}")
```

```python
read_dataset_file("assurance")
```

```python
for datasetfile in datasetsdf["DatasetFile"].unique():
    print(datasetfile)
    datasetdf = read_dataset_file(datasetfile)
    datasetdf = datasetdf.sample(frac=1).reset_index(drop=True)
    datasetdf.to_feather(rootdir / "datasets" / (datasetfile+".dataset.feather2"))
    print("OK")
```

```python
extractionfile,website = datasetsdf.iloc[0][["ExtractionFile","Website"]]
urlsfile = rootdir / "extraction" / (extractionfile+".urls.dataframe.feather")
urlsdf = pd.read_feather(urlsfile)
urlsdf
```

```python
urlsdf["Website"] = urlsdf.apply(lambda x: 4, axis=1)
urlsdf = urlsdf[urlsdf.columns[-1:].tolist() + urlsdf.columns[:-1].tolist()]
urlsdf
```

```python
allurlsdf = None
extractionfiles = set()
for idx,row in datasetsdf.iterrows():
    website = row["Website"]
    extractionfile = row["ExtractionFile"]
    if extractionfile in extractionfiles:
        continue
    else:
        extractionfiles.add(extractionfile)
    print(extractionfile)
    urlsfile = rootdir / "extraction" / (extractionfile+".urls.dataframe.feather")
    urlsdf = pd.read_feather(urlsfile)
    urlsdf["Website"] = urlsdf.apply(lambda x: website, axis=1)
    #urlsdf = urlsdf[urlsdf.columns[-1:].tolist() + urlsdf.columns[:-1].tolist()]
    if allurlsdf is None:
        allurlsdf = urlsdf
    else:
        allurlsdf = allurlsdf.append(urlsdf)
```

```python
allurlsdf = allurlsdf.fillna(0)
```
